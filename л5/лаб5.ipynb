# Взвешенный медианный фильтр

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from skimage import io
from PIL import Image
import statistics
```

```python
def median_img(img_array):
  width, height = len(img_array), len(img_array[0])
  for hor in range(width):
    for vert in range(height):
      #смена коэфициента дает изменение матрицы. 1 - 3х3, 2 - 5х5, 3 - 7х7
      k = 2
      if hor >= k and hor<width-k:
        if vert >= k and vert<height-k:
          for i in range(0, 3):
            a=[]
            for j in range(-k, k+1):
              for e in range(-k, k+1):
                a.append(img_array[hor+j][vert+e][i])
            img_array[hor][vert][i] = statistics.median(a)

  return img_array
```

```python
img_arr = np.array(Image.open('vulkan.jpg'))
io.imshow(img_arr)
```

```python
a = median_img(img_arr)
io.imshow(a)
```
